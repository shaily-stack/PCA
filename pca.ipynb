```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data=pd.read_csv('C:/Users/NF_HR/Downloads/creditcard.csv')
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
features=data.iloc[:,data.columns!='Class']
```

```python
scaler= StandardScaler()
scaler.fit(features)
```

```python
scaled_data= scaler.transform(features)
```

```python
scaled_data=pd.DataFrame(scaled_data,columns= features.columns)
```

```python
scaled_data.head()
```

```python
data.describe().round(1)
```

```python
scaled_data.describe().round(1)
```

```python
scaled_data.drop(columns=['Time'],inplace=True)
```

```python
scaled_data.info()
```

```python
from sklearn.decomposition import PCA
```

```python
pca= PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca= pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
print(pca.explained_variance_)
```

###### This small percentage is because Vi are already PCs of original data in our case


```python
plt.figure(figsize=(18,8))
plt.scatter(x_pca[:,0],x_pca[:,1],c=data['Class'],cmap='plasma')
plt.legend(labels=['Normal', 'Fraud'])
plt.show()
```

```python
x_new= pca.inverse_transform(x_pca)
#plot data
plt.scatter(x_pca[:,0],x_pca[:,1],alpha=0.2)
plt.scatter(x_new[:,0],x_new[:,1],alpha=0.8)

# for length, vector in zip(pca.explained_variance_, pca.components_):
#     v= vector * 3 *np.sqrt(length)
#     draw_vector(pca.mean_,pca.mean_+ v)
plt.axis('equal');    
```

```python
x_new= pca.inverse_transform(x_pca)
#plot data
plt.scatter(scaled_data[:,0],scaled_data[:,1],alpha=0.2)
plt.scatter(x_new[:,0],x_new[:,1],alpha=0.8)

# for length, vector in zip(pca.explained_variance_, pca.components_):
#     v= vector * 3 *np.sqrt(length)
#     draw_vector(pca.mean_,pca.mean_+ v)
plt.axis('equal');    
```
